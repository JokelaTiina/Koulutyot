##### Now we've seen how to plot with Matplotlib and data directly. Let's practice using Matplotlib to plot with pandas.

First we'll need to import pandas and create a DataFrame work with.

```python
## Your code here

#Import pandas as pd
import pandas as pd
```

```python
## Your code here

# Import the './bda1_mathplotlib_answers/car-sales.csv' into a DataFame called car_sales and view
car_sales = pd.read_csv('./bda1_mathplotlib_answers/car-sales.csv')
car_sales
```

```python
## Your code here

# Try to plot the 'Price' column using the plot() function
car_sales['Price'].plot()
```

Why doesn't it work?

**Hint:** It's not numeric data.


```python
## Your code here

# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
car_sales['Price'] = car_sales['Price'].str.replace('$', '', regex=True)
car_sales['Price'] = car_sales['Price'].str.replace(',', '', regex=True)
car_sales['Price'] = car_sales['Price'].str.replace('\.00', '', regex=True)

car_sales['Price'] = car_sales['Price'].astype(int)
car_sales
```

```python
car_sales.info()
```

Next create another column which adds the total amount of sales and another one which shows what date the car was sold.

**Hint:** To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

```python
## Your code here
from datetime import date
import datetime

# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column
car_sales['Total Sales'] = car_sales['Price'].cumsum()
# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
car_sales['Sale Date'] = pd.date_range(start=date.today(), end=date.today() + datetime.timedelta(days=9), freq='D')
# View the car_sales DataFrame
car_sales
```

Now let's try comparing two variables versus the target variable.

More specifially we'll see how age and cholesterol combined effect the target in **patients over 50 years old**.

For this next challenge, we're going to be replicating the following plot:

<img src="./bda1_mathplotlib_answers/figure8.png" width=500/>

```python
# Replicate the above plot in whichever way you see fit

# Note: The method below is only one way of doing it, yours might be
# slightly different

## Your code here

# Create DataFrame with patients over 50 years old
df = heart_disease[heart_disease['age']>50]

# Create the plot
fig, ax = plt.subplots(figsize=(8, 5))

# Plot the data 
# hint. use scatter parameter c = target
scatter = ax.scatter(df['age'], df['chol'],  c=heart_disease['target'])
legend = ax.legend(*scatter.legend_elements(),
                    loc="upper right", title="Target")
ax.add_artist(legend)

# Plot the average "chol" line 
ax.axhline(df['chol'].mean(), color='b', linestyle='dashed', linewidth=1)

# Customize the plot xlabel, ylabel, title and add legend
# hint. For scatter legends see:
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/scatter_with_legend.html
ax.set_xlabel('Age')
ax.set_ylabel('Cholesterol')
ax.set_title('Heart Desiase and Cholesterol Levels')
```

Beatiful, now you've created a plot of two different variables, let's change the style.

Let's try with non-binary heart_disease["trestbps"] data. In many applications it is also useful to change plot colormap (='cmap') and add colorbar. So let's try also those.

Completing this step correctly should result in a figure which looks like this:

<img src="./bda1_mathplotlib_answers/figure9.png" width=500 />

```python
# Replicate the above plot in whichever way you see fit

# Note: The method below is only one way of doing it, yours might be
# slightly different

## Your code here

# Create DataFrame with patients over 50 years old
df = heart_disease[heart_disease['age']>50]
# Create the plot
fig, ax = plt.subplots(figsize=(10, 7))

# Plot the data 
# hint. use scatter parameter cmap = 'cool'
im = scatter = ax.scatter(df['age'], df['chol'],  c=heart_disease['trestbps'], cmap = 'cool')

# Plot the average "chol" line 
ax.axhline(df['chol'].mean(), color='b', linestyle='dashed', linewidth=1)

# Customize the plot
ax.set_xlabel('Age')
ax.set_ylabel('Cholesterol')
ax.set_title('Heart Disease and Cholesterol Levels')

# Add colorbar to fig
fig.colorbar(im, ax=ax)
# Add also legend as a reference for colorbar
legend = ax.legend(*scatter.legend_elements(),
                    loc="upper right", title="trestbps")
ax.add_artist(legend)
```
